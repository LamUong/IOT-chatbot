```python
data = []
for element_l in L:
    for pattern in element_l["patterns"]:
        data.append(["",pattern,element_l["intent"][0]+" "+element_l["intent"][1]])

                    
```

```python
labels = []
labels_index = {}
texts = []

print(data[0])

j = 0
for i, item in enumerate(data):
    if item[2] not in labels_index:
        labels_index[item[2]] = j
        j += 1
        
for item in data:
    labels.append(labels_index[item[2]])
    texts.append(item[1])
#     split_str = item[1].split(' ')
    
#     new_str = []
#     for item in split_str:
#         if len(item) > 3:
#             new_str.append(item.lower().strip())
    
#     texts.append(' '.join(new_str))
print(labels)
print(labels_index)
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

#MAX_NB_WORDS = 117
test_index = {}
for i in range(0,len(texts)):
    if texts[i][:2] =="t:":
        test_index[i] = True
print(test_index)
print(labels)
# keras tokenizer, limit number of words to tokenize to MAX_NB_WORDS
tokenizer = Tokenizer(lower=True)
tokenizer.fit_on_texts(texts)

sequences = tokenizer.texts_to_sequences(texts)

# the complete word index
word_index = tokenizer.word_index
print(word_index)
print(sequences)
```

```python
max_len = 0
for seq in sequences:
    if len(seq) > max_len:
        max_len = len(seq)

MAX_SEQUENCE_LENGTH = max_len


    
# stuff in sequences as a list of word indices
train_x = []
train_y = []
test_x = []
test_y = []
for i in range(0,len(data),1):
    if i in test_index:
        test_x.append(sequences[i])
        test_y.append(labels[i])
    else:
        train_x.append(sequences[i])
        train_y.append(labels[i])

# restricting length of sequences to MAX_SEQUENCE_LENGTH
# takes the last MAX_SEQUENCE_LENGTH words in the list
train_x = pad_sequences(train_x, maxlen=MAX_SEQUENCE_LENGTH)
test_x = pad_sequences(test_x, maxlen=MAX_SEQUENCE_LENGTH)
print(sequences)
print(train_x)
print(test_x)
# stuff in sequences after processing


```

```python
import numpy as np

train_y = keras.utils.to_categorical(np.asarray(train_y))
test_y = keras.utils.to_categorical(np.asarray(test_y))

```

```python
# number of docs X size of sequence
print('Shape of data tensor:', train_x.shape)

# number of docs X size of label vector
print('Shape of label tensor:', train_y.shape)

print('Shape of data tensor:', test_x.shape)

# number of docs X size of label vector
print('Shape of label tensor:', test_y.shape)
```

```python


x_train = train_x
y_train = train_y
x_val = test_x
y_val = test_y
```

```python
import os

embeddings_index = {}
GLOVE_DIR = 'data'

f = open(os.path.join(GLOVE_DIR, 'glove.6B.300d.txt'))
for i, line in enumerate(f):
    values = line.split(' ')
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()
print('Found %s word vectors.' % len(embeddings_index))
```

```python
EMBEDDING_DIM = 300

# set up a matrix to store the embedding vector for each word in word_index
embedding_matrix = np.zeros((len(word_index), EMBEDDING_DIM))
print(embedding_matrix.shape)

for i, word in enumerate(word_index):
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector
        
```

```python
from keras.layers import Embedding, Input, Dense, Conv1D, MaxPooling1D, Flatten, AveragePooling1D
from keras.layers.pooling import GlobalMaxPooling1D
from keras.layers.core import Dropout
from keras.models import Model
from keras.layers import Conv2D, MaxPooling2D, Input, Dense, add, Activation, concatenate, Flatten

embedding_layer = Embedding(len(word_index),
                            EMBEDDING_DIM,
                            weights=[embedding_matrix],
                            input_length=MAX_SEQUENCE_LENGTH,
                            trainable=True)

sequence_input = Input(shape=(MAX_SEQUENCE_LENGTH,), dtype='int32')

embedded_sequences = embedding_layer(sequence_input)

x1 = Conv1D(50, 2, activation='relu',padding="same", strides=1)(embedded_sequences)
x2 = Conv1D(50, 5, activation='relu',padding="same", strides=1)(embedded_sequences)
x3 = Conv1D(50, 10, activation='relu',padding="same", strides=1)(embedded_sequences)

x = concatenate([x1, x2 ,x3],  axis=-1)
x = Activation("relu")(x)
x = Dropout(0.2)(x)

x = AveragePooling1D(2)(x)

x = Conv1D(512, 2, activation='relu',padding="same", strides=1)(x)
x = AveragePooling1D(2)(x)

x = Flatten()(x)
x = Dense(64, activation='relu')(x)
x = Dropout(0.4)(x)
x = Dense(32, activation='relu')(x)
x = Dropout(0.3)(x)

preds = Dense(len(labels_index), activation='softmax')(x)

model = Model(sequence_input, preds)
model.summary()
model.compile(loss='categorical_crossentropy',
              optimizer=keras.optimizers.RMSprop(lr=0.0001, rho=0.4, epsilon=1e-08, decay=0.0),
              metrics=['acc'])

model.fit(x_train, y_train, validation_data=(x_val, y_val),
          epochs=10000)
```
